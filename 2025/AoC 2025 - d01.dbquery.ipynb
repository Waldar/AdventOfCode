```sql
with cte_data (pos, val, mod) as
(
select posexplode(array(strt) || transform(split(translate(value, 'LR', '-+'), '\n'), t -> t::int)) as (pos, val)
     , mod
  from values(50, 100) as (strt, mod)
  join read_files('/Volumes/waldar/fabien/aoc2025/aoc2025_01_input.txt', format => 'text', wholeText => true) as t (value) on true
/*join values( 'L68' || '\n'
            || 'L30' || '\n'
            || 'R48' || '\n'
            || 'L5'  || '\n'
            || 'R60' || '\n'
            || 'L55' || '\n'
            || 'L1'  || '\n'
            || 'L99' || '\n'
            || 'R14' || '\n'
            || 'L82') as t (value) on true*/
)
  ,  cte_rotation (rot, clicks) as
(
select pmod(sum(val) over W1, mod) as rot
     , case
         when val > 0
         then floor((pmod(sum(val) over W2, mod) + val    ) / mod)
         else floor((pmod(sum(val) over W2, mod)       - 1) / mod)
            - floor((pmod(sum(val) over W2, mod) + val - 1) / mod)
       end as clicks
  from cte_data
window W1 as (order by pos asc)
     , W2 as (order by pos asc rows between unbounded preceding and 1 preceding)
)
select count(*) filter(where rot = 0) as part1
     , sum(clicks)                    as part2
     , decode(x'434C49434B', 'utf-8') as method_part2
  from cte_rotation;
```
